```python
# Core libraries
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn - Data Processing & Modeling
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, recall_score, roc_auc_score,
    matthews_corrcoef, confusion_matrix
)

from sklearn.ensemble import (
    RandomForestClassifier, ExtraTreesClassifier,
    GradientBoostingClassifier, AdaBoostClassifier
)
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier

# Gradient boosting libraries
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
```

```python
df = pd.read_excel('/content/hemo_data_complete.xlsx')
```

```python
# Features and labels
X = df.drop(columns=["ID", "seq", "Î¼M", "label"])
y = df['label']
```

```python

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Models
models = {
    'LightGBM': LGBMClassifier(),
    'RandomForest': RandomForestClassifier(),
    'ExtraTrees': ExtraTreesClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'GradientBoosting': GradientBoostingClassifier(),
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'AdaBoost': AdaBoostClassifier(),
    'DecisionTree': DecisionTreeClassifier(),
    'RidgeClassifier': RidgeClassifier()
}

# Metric function
def basic_metrics(y_true, y_pred, y_prob):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sens = tp / (tp + fn)
    spec = tn / (tn + fp)
    acc = accuracy_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob)
    mcc = matthews_corrcoef(y_true, y_pred)
    return {
        'Sens(%)': round(sens * 100, 2),
        'Spec(%)': round(spec * 100, 2),
        'Acc(%)': round(acc * 100, 2),
        'AUC': round(auc, 4),
        'MCC': round(mcc, 4)
    }

# Evaluation loop
all_results = []

for name, model in models.items():
    print(f"Training: {name}")
    start = time.time()
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    cv_metrics = []
    for train_idx, val_idx in skf.split(X_train, y_train):
        X_tr, X_val = X_train[train_idx], X_train[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        model.fit(X_tr, y_tr)
        y_val_pred = model.predict(X_val)
        y_val_prob = model.predict_proba(X_val)[:, 1] if hasattr(model, "predict_proba") else y_val_pred
        cv_metrics.append(basic_metrics(y_val, y_val_pred, y_val_prob))

    # Average and std of CV metrics
    cv_df = pd.DataFrame(cv_metrics)
    cv_avg = cv_df.mean().to_dict()
    cv_std = cv_df.std().to_dict()

    cv_avg = {f"Train_{k}": round(v, 4) for k, v in cv_avg.items()}
    cv_std = {f"Train_{k}_std": round(v, 4) for k, v in cv_std.items()}

    # Test set metrics
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    y_test_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_test_pred
    test_scores = basic_metrics(y_test, y_test_pred, y_test_prob)
    test_scores = {f"Test_{k}": v for k, v in test_scores.items()}

    # Combine all
    result = {'Model': name, 'Time(s)': round(time.time() - start, 2)}
    result.update(cv_avg)
    result.update(cv_std)
    result.update(test_scores)
    all_results.append(result)

# Save results
results_df = pd.DataFrame(all_results)
results_df.sort_values(by='Test_MCC', ascending=False, inplace=True)
results_df.to_excel("model_metrics_summary.xlsx", index=False)

# Show results
print(results_df)

# Plot Test MCC
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x='Model', y='Test_MCC', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title("Test MCC by Model")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, matthews_corrcoef
import time
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

# Scale features
X_scaled = StandardScaler().fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Enhanced models configuration
models = {
    'LightGBM': LGBMClassifier(
        random_state=42,
        verbose=-1,
        first_metric_only=True  # Use only first metric for early stopping [1]
    ),
    'LightGBM_tuned': LGBMClassifier(
        learning_rate=0.1,
        n_estimators=500,
        num_leaves=31,
        max_depth=10,
        min_child_samples=100,
        feature_fraction=0.8,
        random_state=42,
        verbose=-1,
        first_metric_only=True
    ),  # Based on tuning recommendations [4]
    'RandomForest': RandomForestClassifier(random_state=42),
    'ExtraTrees': ExtraTreesClassifier(random_state=42),
    'XGBoost': XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    ),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'RidgeClassifier': RidgeClassifier(random_state=42)
}

# Metric function with enhanced robustness
def basic_metrics(y_true, y_pred, y_prob):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sens = tp / (tp + fn)  # Sensitivity
    spec = tn / (tn + fp)  # Specificity
    acc = accuracy_score(y_true, y_pred)

    try:
        auc = roc_auc_score(y_true, y_prob)
    except:
        auc = np.nan

    mcc = matthews_corrcoef(y_true, y_pred)

    return {
        'Sens(%)': round(sens * 100, 2),
        'Spec(%)': round(spec * 100, 2),
        'Acc(%)': round(acc * 100, 2),
        'AUC': round(auc, 4) if not np.isnan(auc) else np.nan,
        'MCC': round(mcc, 4)
    }

# Evaluation loop with improvements
all_results = []

for name, model in models.items():
    print(f"Training: {name}")
    start = time.time()
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    cv_metrics = []
    for train_idx, val_idx in skf.split(X_train, y_train):
        X_tr, X_val = X_train[train_idx], X_train[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        model.fit(X_tr, y_tr)
        y_val_pred = model.predict(X_val)

        # Handle probability/score extraction robustly
        if hasattr(model, "predict_proba"):
            y_val_prob = model.predict_proba(X_val)[:, 1]
        elif hasattr(model, "decision_function"):
            y_val_prob = model.decision_function(X_val)
        else:
            y_val_prob = y_val_pred

        cv_metrics.append(basic_metrics(y_val, y_val_pred, y_val_prob))

    # Average CV metrics
    cv_avg = pd.DataFrame(cv_metrics).mean().to_dict()
    cv_avg = {f"Train_{k}": v for k, v in cv_avg.items()}

    # Test set evaluation
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_test_prob = model.predict_proba(X_test)[:, 1]
    elif hasattr(model, "decision_function"):
        y_test_prob = model.decision_function(X_test)
    else:
        y_test_prob = y_test_pred

    test_scores = basic_metrics(y_test, y_test_pred, y_test_prob)
    test_scores = {f"Test_{k}": v for k, v in test_scores.items()}

    # Combine results
    result = {'Model': name, 'Time(s)': round(time.time() - start, 2)}
    result.update(cv_avg)
    result.update(test_scores)
    all_results.append(result)

# Save and display results
results_df = pd.DataFrame(all_results)
results_df.to_excel("model_metrics_summary.xlsx", index=False)
results_df
```
